## 1. Bucles y Ciclos

Bucle for El bucle for se utiliza para iterar sobre una secuencia (como una lista, tupla, diccionario, conjunto o cadena de caracteres). Se ejecuta una vez por cada elemento en la secuencia.

image.png

```python
# Imprimir todos los números de una secuencia de datos.

secuencia = [1,2,3,4,5,]
# La "i" es tomada como la variable de nuestra secuencia (lista de datos) que se irán tomando hasta que se termine.
 for i in secuencia
  print(i)
```
