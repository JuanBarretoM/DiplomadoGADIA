## REGRESIÓN LINEAL

```python
# Gráfica de dispersión: salario vs experiencia.
# La primera parte es del eje X y la segunda del eje Y.

Y = salarios["Salario"]
X = salarios["Años de Experiencia"]

# "scatter" es un tipo de gráfica de dispersión.
# El primer valor es el eje X y el segundo valor es el eje Y para poder generar la gráfica.

plt.scatter(X,Y,color = "red")

# Título del gráfico.
plt.title("Salario vs Años de Experiencia")

# Título eje X.
plt.xlabel("Años de Experiencia")

# Título eje Y.
plt.ylabel("Salarios")

# Mostrar Gráfica.
plt.show()
```
